## scikit-learn
## y_pred = X_ex_train * w  + b

```python
from sklearn.linear_model import LinearRegression
```

```python
# scale
slr = LinearRegression()
slr.fit(X_ex_train, y_ex_train)
y_pred = slr.predict(X_ex_train)
print('Slope: %.3f' % slr.coef_[0])
print('Intercept: %.3f' % slr.intercept_)
```

```python
def lin_regplot(X, y, model):
    plt.scatter(X, y, c='lightblue')
    plt.plot(X, model.predict(X), color='red', linewidth=2)    
    return
# (x,y)(X, model.predict(X))
```

```python
lin_regplot(X_ex_train, y_ex_train, slr)

plt.xlabel('sqft_living')
plt.ylabel('Price')
#plt.tight_layout()
plt.show()
```

```python
slr.predict(1000)
```

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error

print('MAE train: %.3f' % (
        mean_absolute_error(y_ex_train, y_pred)))
print('MSE train: %.3f' % (
        mean_squared_error(y_ex_train, y_pred)))
print('R^2 train: %.3f' % (
        r2_score(y_ex_train, y_pred)))
```

```python
ex_train = np.sort(ex_train, axis=1, kind='mergesort')
```

```python
ex_train
```

```python
ex_train.shape
```

```python
X_ex_train_sort = ex_train[0,:][:, np.newaxis]
y_ex_train_sort = ex_train[1,:]
```

https://docs.scipy.org/doc/numpy-1.8.1/reference/generated/numpy.sort.html#numpy.sort

```python
from sklearn.preprocessing import PolynomialFeatures

lr = LinearRegression()
pr = LinearRegression()

# 
quadratic = PolynomialFeatures(degree=2)

X_quad = quadratic.fit_transform(X_ex_train_sort)
```

http://scikit-learn.org/stable/auto_examples/linear_model/plot_polynomial_interpolation.html#sphx-glr-auto-examples-linear-model-plot-polynomial-interpolation-py

```python
X_quad
```

```python
pr.fit(X_quad, y_ex_train_sort)

y_quad_fit = pr.predict(X_quad)
```

```python
y_quad_fit
```

```python
# fit linear features
lr.fit(X_ex_train_sort, y_ex_train_sort)
y_lin_fit = lr.predict(X_ex_train_sort)

# fit quadratic features
pr.fit(X_quad, y_ex_train_sort)
y_quad_fit = pr.predict(quadratic.fit_transform(X_ex_train_sort))

# plot results
plt.scatter(X_ex_train_sort, y_ex_train_sort, label='training points')

plt.plot(X_ex_train_sort, y_lin_fit, label='linear fit', linestyle='--')
plt.plot(X_ex_train_sort, y_quad_fit, label='quadratic fit')
plt.legend(loc='upper left')

plt.xlabel('sqft_living')
plt.ylabel('Price')

plt.show()
```

```python
print('MAE train: %.3f' % (
        mean_absolute_error(y_ex_train_sort, y_quad_fit)))
print('MSE train: %.3f' % (
        mean_squared_error(y_ex_train_sort, y_quad_fit)))
print('R^2 train: %.3f' % (
        r2_score(y_ex_train_sort, y_quad_fit)))
```

```python
#      cube four five 

quadratic = PolynomialFeatures(degree=2)
cube = PolynomialFeatures(degree=3)
four = PolynomialFeatures(degree=4)
five = PolynomialFeatures(degree=5)

X_quad = quadratic.fit_transform(X_ex_train_sort)
X_cube = cube.fit_transform(X_ex_train_sort)
X_four = four.fit_transform(X_ex_train_sort)
X_five = five.fit_transform(X_ex_train_sort)

cr = LinearRegression()
fourr = LinearRegression()
fiver = LinearRegression()

pr.fit(X_quad, y_ex_train_sort)
cr.fit(X_cube, y_ex_train_sort)
fourr.fit(X_four, y_ex_train_sort)
fiver.fit(X_five, y_ex_train_sort)

y_quad_fit = pr.predict(X_quad)
y_cube_fit = cr.predict(X_cube)
y_four_fit = fourr.predict(X_four)
y_five_fit = fiver.predict(X_five)
```
